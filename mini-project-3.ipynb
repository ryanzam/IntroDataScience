```python
#imports
from keras.layers import Dense
from keras.models import Sequential
from matplotlib import pyplot as plt
from sklearn.model_selection import train_test_split
from matplotlib import pyplot as plt

import numpy as np
import pandas as pd


#loading housing dataset
df = pd.read_csv("housing_data.csv", delim_whitespace = True, header=None)
dataset = df.values

# spliting dataset into x as input and y as output variables
x = dataset[:,0:13]
y = dataset[:,13]

# splitting data into training set and test set - 80%-20% for this project
(X_train, X_test, Y_train, Y_test) = train_test_split(x, y, test_size = 0.2)


model = Sequential()
model.add(Dense(13, input_dim=13, kernel_initializer='normal', activation='relu'))
model.add(Dense(6, kernel_initializer='normal', activation='relu'))
model.add(Dense(1, kernel_initializer='normal'))

# Compile model
model.compile(loss='mean_squared_error', optimizer='adam', metrics=['accuracy'])

# fit the model
history = model.fit(X_train, Y_train, validation_data=(X_test, Y_test), epochs=100, batch_size=5)

#evaluate the model
scores = model.evaluate(X_test, Y_test)
print("\n%s: %.2f%%" % (model.metrics_names[1], scores[1]*100))

#plots
plt.plot(history.history['acc'])
plt.plot(history.history['val_acc'])
plt.title('model accuracy')
plt.ylabel('accuracy')
plt.xlabel('epoch')
plt.legend(['train', 'test'], loc='upper left')
plt.show()
# summarize history for loss
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.title('model loss')
plt.ylabel('loss')
plt.xlabel('epoch')
plt.legend(['train', 'test'], loc='upper left')
plt.show()
```
